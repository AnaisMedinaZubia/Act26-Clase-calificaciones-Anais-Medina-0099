```python
class Calificacion:
    def __init__(self):
        # Atributos (Datos referentes a la clase)
        self.alumno = ""
        self.materia = ""
        self.nota = 0.0

    def leer_datos(self, numero_objeto):
        """Función para capturar la información del usuario"""
        print(f"\n--- Ingresando datos para el Objeto {numero_objeto} ---")
        self.alumno = input("Nombre del alumno: ")
        self.materia = input("Nombre de la materia: ")
        # Validamos que la nota sea un número

    def mostrar_datos(self):
        """Función para imprimir la información en pantalla"""
        print("\n===============================")
        print(f"RESUMEN DE CALIFICACIÓN")
        print(f"Alumno:  {self.alumno}")
        print(f"Materia: {self.materia}")
        print(f"Nota:    {self.nota}")

        # Mensaje personalizado según la nota
        if self.nota >= 60:
            print("Estado:  Aprobado ✅")
        else:
            print("Estado:  Reprobado ❌")
        print("===============================")

# --- Creación de los Objetos (Instancias de la clase) ---

# Creamos el primer objeto
calificacion1 = Calificacion()
calificacion1.leer_datos("", "", "")


# Creamos el segundo objeto
calificacion2 = Calificacion()
calificacion2.leer_datos(2)

# --- Ejecución de las funciones de los objetos ---
print("\n--- MOSTRANDO RESULTADOS FINALES ---")
calificacion1.mostrar_datos()
calificacion2.mostrar_datos()
print("prograa realizado por Anais medina NL = 36")
```
